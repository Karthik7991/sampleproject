```
import numpy as np
import random 
import matplotlib.pyplot as plt
from statistics import *
```

```
s_z = [50,100,150,200,250,300]                             # different sample sizes and 't' is taken as the index number for list s_z
t = 0                                                      

rows, cols = 3, 2                                          # we have arranged the plots in rows and columns
fig, pic = plt.subplots(rows, cols, figsize=(20,15))      
for i in range(0, rows):
  for j in range(0, cols):
                                                                                            
    y = []                                                 # here for all plots we have taken 3000 trails and sample size for each plot varies
    for l in range(0,3000):                                
      x = []
      for n in range(0,s_z[t]):                            # first we generate s_z[t] random numbers and appended them to list 'x' 
        value = random.randint(-20,20)                     # then we find mean of list 'x' and append to list 'y' and through by iteration 
        x.append(value)                                    # we generate 3000 sample means to y 
      k = mean(x) 
      y.append(k)

    pic[i, j].hist(y)                                           # ploting histogram diagram
    pic[i, j].set_title(label="sample size :" + str(s_z[t]))    # title to each plot
    t += 1                                                      # increasing t value by 1 for further plot
    pic[i, j].set_xlabel("mean of samples")                     
    pic[i, j].set_ylabel("distributions")
    pic[i, j].set_xlim(-5,5)
    pic[i, j].set_ylim(0,1000)      
plt.show()
plt.savefig('clt.png')
```
